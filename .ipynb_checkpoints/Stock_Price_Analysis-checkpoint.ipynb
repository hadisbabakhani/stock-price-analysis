```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def fetch_stock_data(symbol, api_key):
    """Fetch daily stock data from Alpha Vantage API"""
    url = 'https://www.alphavantage.co/query'
    params = {
        'function': 'TIME_SERIES_DAILY',
        'symbol': symbol,
        'apikey': api_key,
        'outputsize': 'compact'
    }

    response = requests.get(url, params=params)

    if response.status_code != 200:
        raise Exception(f"Error fetching data: {response.status_code}")

    data = response.json()

    # Check for error message in the response
    if 'Error Message' in data:
        raise ValueError(f"Error in API response: {data['Error Message']}")

    return data

def process_stock_data(data):
    """Convert API response to a Pandas DataFrame"""
    time_series = data.get('Time Series (Daily)', {})

    # Check if time_series is empty
    if not time_series:
        raise ValueError("No data found for the given stock symbol or API key issue.")

    # Convert to DataFrame
    df = pd.DataFrame.from_dict(time_series, orient='index')
    
    # Check if DataFrame has at least 5 columns
    if df.shape[1] < 5:
        raise ValueError("The API response does not contain the expected data structure.")
    
    # Rename columns for better readability
    df.columns = ['open', 'high', 'low', 'close', 'volume']
    
    # Convert the index to datetime and the columns to numeric
    df.index = pd.to_datetime(df.index)
    df = df.astype(float)
    
    return df

def calculate_daily_returns(df):
    """Calculate daily returns as a percentage change"""
    df['daily_return'] = df['close'].pct_change()
    return df

def calculate_moving_averages(df, short_window=20, long_window=50):
    """Calculate short-term and long-term moving averages"""
    df['short_mavg'] = df['close'].rolling(window=short_window).mean()
    df['long_mavg'] = df['close'].rolling(window=long_window).mean()
    return df

def calculate_volatility(df):
    """Calculate stock volatility (standard deviation of returns)"""
    df['volatility'] = df['daily_return'].rolling(window=20).std()
    return df

def calculate_cumulative_returns(df):
    """Calculate cumulative returns"""
    df['cumulative_return'] = (1 + df['daily_return']).cumprod() - 1
    return df

def calculate_average_price(df):
    """Calculate the average closing price"""
    return df['close'].mean()

def plot_stock_data(df, symbol):
    """Plot the stock's closing price and moving averages"""
    plt.figure(figsize=(10, 5))
    plt.plot(df['close'], label='Close Price', color='blue')
    plt.plot(df['short_mavg'], label='20-Day Moving Average', color='orange')
    plt.plot(df['long_mavg'], label='50-Day Moving Average', color='green')
    plt.title(f'{symbol} Prices & Moving Averages')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
    
    plt.figure(figsize=(10, 5))
    df['cumulative_return'].plot(title=f'{symbol} Cumulative Return', ylabel='Cumulative Return')
    plt.show()

# Main script execution
api_key = 'QJMVCCAI6NH0L6CY'  
stock_symbol = 'AAPL'  

try:
    # Fetch stock data
    data = fetch_stock_data(stock_symbol, api_key)
    print("API Response:", data)  # Print the API response

    # Process the data into a DataFrame
    stock_df = process_stock_data(data)

    # Perform various analyses
    stock_df = calculate_daily_returns(stock_df)
    stock_df = calculate_moving_averages(stock_df, short_window=20, long_window=50)
    stock_df = calculate_volatility(stock_df)
    stock_df = calculate_cumulative_returns(stock_df)

    # Display the first few rows of the DataFrame
    print(stock_df.head())

    # Calculate the average closing price
    average_price = calculate_average_price(stock_df)
    print(f'Average Closing Price for {stock_symbol}: ${average_price:.2f}')

    # Show the calculated volatility (standard deviation)
    latest_volatility = stock_df['volatility'].iloc[-1]
    print(f'Latest Volatility (20-day rolling): {latest_volatility:.4f}')

    # Plot the results
    plot_stock_data(stock_df, stock_symbol)

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
def calculate_rsi(df, window=14):
    """Calculate the Relative Strength Index (RSI)"""
    delta = df['close'].diff(1)
    gain = np.where(delta > 0, delta, 0)
    loss = np.where(delta < 0, -delta, 0)
    
    avg_gain = pd.Series(gain).rolling(window=window, min_periods=1).mean()
    avg_loss = pd.Series(loss).rolling(window=window, min_periods=1).mean()
    
    rs = avg_gain / avg_loss
    df['rsi'] = 100 - (100 / (1 + rs))
    
    return df

def calculate_bollinger_bands(df, window=20):
    """Calculate Bollinger Bands"""
    df['rolling_mean'] = df['close'].rolling(window=window).mean()
    df['rolling_std'] = df['close'].rolling(window=window).std()
    
    df['upper_band'] = df['rolling_mean'] + (df['rolling_std'] * 2)
    df['lower_band'] = df['rolling_mean'] - (df['rolling_std'] * 2)
    
    return df

# Calculate RSI and Bollinger Bands
stock_df = calculate_rsi(stock_df)
stock_df = calculate_bollinger_bands(stock_df)

# Show the updated DataFrame with RSI and Bollinger Bands
print(stock_df[['close', 'rsi', 'upper_band', 'lower_band']].head())

# Plot the Bollinger Bands
stock_df[['close', 'upper_band', 'lower_band']].plot(title=f'{stock_symbol} Bollinger Bands', figsize=(10, 5), ylabel='Price')

# Plot RSI
stock_df['rsi'].plot(title=f'{stock_symbol} Relative Strength Index (RSI)', figsize=(10, 5), ylabel='RSI')
```

```python
# Final Summary of Key Metrics

# 1. Average Closing Price
average_price = calculate_average_price(stock_df)
print(f'Final Average Closing Price for {stock_symbol}: ${average_price:.2f}')

# 2. Latest Volatility (Standard Deviation of Daily Returns over 20 days)
latest_volatility = stock_df['volatility'].iloc[-1]
print(f'Final Latest Volatility (20-day rolling): {latest_volatility:.4f}')

# 3. Cumulative Return over the Period
final_cumulative_return = stock_df['cumulative_return'].iloc[-1]
print(f'Final Cumulative Return for {stock_symbol}: {final_cumulative_return:.2%}')

# 4. RSI Analysis
rsi_value = stock_df['rsi'].iloc[-1]
print(f'Final RSI (Relative Strength Index): {rsi_value:.2f}')
if rsi_value > 70:
    print(f'{stock_symbol} is in overbought territory (RSI > 70).')
elif rsi_value < 30:
    print(f'{stock_symbol} is in oversold territory (RSI < 30).')
else:
    print(f'{stock_symbol} RSI is neutral (30 < RSI < 70).')

# 5. Bollinger Bands (Latest Close vs. Upper/Lower Bands)
upper_band = stock_df['upper_band'].iloc[-1]
lower_band = stock_df['lower_band'].iloc[-1]
latest_close = stock_df['close'].iloc[-1]

print(f'Final Closing Price: ${latest_close:.2f}')
print(f'Bollinger Bands: Upper Band = ${upper_band:.2f}, Lower Band = ${lower_band:.2f}')
if latest_close > upper_band:
    print(f'{stock_symbol} is trading above the upper Bollinger Band (potential overbought condition).')
elif latest_close < lower_band:
    print(f'{stock_symbol} is trading below the lower Bollinger Band (potential oversold condition).')
else:
    print(f'{stock_symbol} is trading within the Bollinger Bands (normal range).')

# Plot all visualizations

# Plot closing price, short-term and long-term moving averages
stock_df[['close', 'short_mavg', 'long_mavg']].plot(title=f'{stock_symbol} Prices & Moving Averages', figsize=(10, 5), ylabel='Price')

# Plot Bollinger Bands
stock_df[['close', 'upper_band', 'lower_band']].plot(title=f'{stock_symbol} Bollinger Bands', figsize=(10, 5), ylabel='Price')

# Plot Cumulative Return
stock_df['cumulative_return'].plot(title=f'{stock_symbol} Cumulative Return', figsize=(10, 5), ylabel='Cumulative Return')

# Plot RSI
stock_df['rsi'].plot(title=f'{stock_symbol} Relative Strength Index (RSI)', figsize=(10, 5), ylabel='RSI')
```
